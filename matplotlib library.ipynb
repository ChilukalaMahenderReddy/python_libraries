```python
#importing matplotlib library
from matplotlib import pyplot as plt
```

```python
#creating listof values for X and Y axis 
x=[2,6,4]
y=[10,15,5]
#ploting for x and y list
plt.plot(x,y)
plt.show()
```

```python
#addign extra properties to the plot
x1=[2,4,6]
y1=[10,15,12]
x2=[4,6,8]
y2=[8,25,3]
#adding the line color and labels
plt.plot(x1,y1,'g',label='list1')
plt.plot(x2,y2,'o:r',marker='*',label='list2')
#ploting the title and x, y axis labels
plt.title("matplotlib example")
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
#adding labels to the plot
plt.legend()
#adding gridview to plot
plt.grid()
plt.show()
```

```python
#ploting the cars price using bar plot
car_name=['Tata','Mahindra','Audi','BMW','Skoda','Toyota']
car_price=[14,15,30,25,18,20]#price in lacks
plt.bar(car_name,car_price,color='green')
plt.xlabel('Car Names')
plt.ylabel('Car Prices')
plt.title('Comparision of different cars price')
plt.show()
```

```python
#ploting the cars price using multiple bar plots
car_name=['Tata','Mahindra','Audi','BMW','Skoda','Toyota']
car_price=[14,15,30,25,18,20]#price in lacks

car_name1=['Tata1','Mahindra1','Audi1','BMW1','Skoda1','Toyota1']
car_price1=[20,10,25,15,10,20]#price in lacks
#adding figure size to plot
plt.figure(figsize=(10,5))
plt.bar(car_name,car_price,label='first',width=.5)
plt.bar(car_name1,car_price1,color='red',label='second',width=.5)
plt.xlabel('Car Names')
plt.ylabel('Car Prices')
plt.legend()
plt.title('Comparision of different cars price')
plt.show()
```

```python
#scatter plot
plt.scatter([1,2,3,4,5],[4,3,6,8,2],color='red')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.title('Scatter Plot')
plt.show()
```

```python
#displaying multiple plots at the same time
#plot 1:
x = [0, 1, 2, 3]
y = [3, 8, 1, 10]

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = [0, 1, 2, 3]
y = [10, 20, 30, 40]

plt.subplot(1, 2, 2)
plt.plot(x,y,'r')

plt.show()
```

```python
#histogram is used to dispaly freequency disrtibution
import numpy as np

x=np.random.randint(10,90,50)
plt.hist(x)
plt.show()
```

```python
#pie chart
y = np.array([35, 25, 25, 15])

plt.pie(y,labels=['a','b','c','d'],explode=[0.0,0.1,0.0,0.0],shadow=True)
plt.title('Pie Chart')
plt.legend()
plt.show()
```

```python
#displaing area plot
days = [1,2,3,4,5]
sleeping =[7,8,6,11,7]
eating = [2,3,4,3,2]
working =[7,8,7,2,2]
playing = [8,5,7,8,13]
plt.figure(figsize=(8,5))
plt.plot([],[],color='m', label='Sleeping', linewidth=5)
plt.plot([],[],color='c', label='Eating', linewidth=5)
plt.plot([],[],color='r', label='Working', linewidth=5)
plt.plot([],[],color='k', label='Playing', linewidth=5)
  
plt.stackplot(days, sleeping,eating,working,playing, colors=['m','c','r','k'])
  
plt.xlabel('x')
plt.ylabel('y')
plt.title('Stack Plot')
plt.legend()
plt.show()
```
