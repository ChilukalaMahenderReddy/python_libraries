```python
#OpenCV library is used for the image operations
#if OpenCV library is not installed in your system use below command
#!pip install opencv-python
import cv2 as cv
import numpy as np
```

```python
#OpenCV library supports following formats
'''
Windows bitmaps – *.bmp, *.dib
JPEG files – *.jpeg, *.jpg
Portable Network Graphics – *.png 
WebP – *.webp 
Sun rasters – *.sr, *.ras 
TIFF files – *.tiff, *.tif 
'''
```

```python
#imread() methos is used for reading the images
#in this method first parameter is image path 
#second parameter by default is 1, 1 indicates that to read image in BGR colore
img = cv.imread("sample.jpg",1)
```

```python
#im.size is used to get the size of an image
print(img.size)
```

```python
#im.shape is used to get the height and width of the image
print(img.shape)
```

```python
#followling 3 methods are used to displaying the image in new window
cv.imshow("image",img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#displaying the images using matplotlib library
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
# Converting BGR color to RGB color format
RGB_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```

```python
#second parameter 0 indicated that read image in grayscale
img1 = cv.imread('sample.jpg',0)
```

```python
#to draw a line in image
#cv2.line(image_path,start_point,end_point,color_RGB,thicknes)
img2 = cv.line(img,(0,0),(300,300),(255,0,0),5)
plt.imshow(img2)
```

```python
# Creating a black screen image using nupy.zeros function
Img = np.zeros((512, 512, 3), dtype='uint8')
img3 = cv.line(Img,(0,0),(300,300),(255,0,0),5)
plt.imshow(img3)
```

```python
#adding eclipse to image
Img1 = np.zeros((512, 512, 3), dtype='uint8')
#image = cv2.ellipse(image, center_coordinates, axesLength,angle, startAngle, endAngle, color, thickness)
image = cv.ellipse(Img1, (200,200), (100,50),0, 0, 360, (0,255,0), 5)
plt.imshow(image)
```

```python
#adding circle to image
image1 = cv.circle(img, (300,300), 100, (20,20,20), 5)
plt.imshow(image1)
```

```python
#adding text to image
#image = cv2.putText(image, 'OpenCV', org, font,fontScale, color, thickness, cv2.LINE_AA)
im = cv.putText(img,'EnterPi',(100,100),cv.FONT_HERSHEY_COMPLEX,2,(255,0,0),3)
plt.imshow(im)
```

```python
#image resizing 
resize_image = cv.resize(im,(300,300))
plt.imshow(resize_image)
```

```python
'''
The basic Thresholding technique is Binary Thresholding. 
For every pixel, the same threshold value is applied. 
If the pixel value is smaller than the threshold, 
it is set to 0, otherwise, it is set to a maximum value.
'''
```

```python
#threshold techniques
'''
cv2.THRESH_BINARY: If pixel intensity is greater than the set threshold, value set to 255, else set to 0 (black).
cv2.THRESH_BINARY_INV: Inverted or Opposite case of cv2.THRESH_BINARY.
cv.THRESH_TRUNC: If pixel intensity value is greater than threshold, it is truncated to the threshold. The pixel values are set to be the same as the threshold. All other values remain the same.
cv.THRESH_TOZERO: Pixel intensity is set to 0, for all the pixels intensity, less than the threshold value.
cv.THRESH_TOZERO_INV: Inverted or Opposite case of cv2.THRESH_TOZERO.'''
```

```python
#cv.threshold(image_source,threshold_value,maximum_vaue,threshold_technique)
ret,thresh1 = cv.threshold(im, 120, 255, cv.THRESH_BINARY)
plt.imshow(thresh1)
```
